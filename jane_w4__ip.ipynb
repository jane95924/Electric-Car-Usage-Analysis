Research Question

i)Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```
#making sure the city is paris
df[df.City == 'Paris'] 
```

```
# select the city in the city column as paris
paris=df.copy()

paris.drop(paris[paris.City != 'Paris'].index, inplace=True) 
```

```
#sort the given data by station and hour 
paris.sort_values(['ID','day','hour','minute'], ascending=True, inplace=True)
paris.head(20)
```

```
#Picking up= negative change in cars at station 
#create column with rate of change of bluecar per minute 
paris['change']= paris['Bluecar counter'].diff()
```

```
#Determine hour with the highest rate of negative change in cars
paris1= paris[paris.change < 0]
paris1.groupby(['hour'])['change'].sum().sort_values(ascending=True)
#most popular hour is hour 21, with total pickups of 111 cars in the period of April 1- Apil 9
```

2)What is the most popular hour for returning cars?

```
#sort the dataset(all cities) by station and hour 
df.sort_values(['ID','day','hour','minute'], ascending=True, inplace=True)
df.head()
```

```
#create column with rate of change of cars per minute 
df['change']= df['Cars'].diff()
```

```
#Returns result in increase of cars at the station therefore:
# Determine hour with the highest rate of positive change in cars
df= df[df.change > 0]
highest = df.groupby(['hour'])['change'].sum().sort_values(ascending=False)
highest
# hour 3 has the highest returns at 219
```
